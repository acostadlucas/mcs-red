```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
params  = {
    'market_conditions': {
        'mean':50,
        'std':15
    }
}
```

```python
def generate_market_conditions(n_simulations, params):
    market_strength = np.random.normal(
        params['market_conditions']['mean'],
        params['market_conditions']['std'],
        n_simulations
    )
    return np.clip(market_strength, 0, 100)
```

```python
# Generate data
n_simulations = 1000
market_data = generate_market_conditions(n_simulations, params)
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.hist(market_data, bins=30, color='skyblue', edgecolor='black')
plt.title('Simulated Market Conditions')
plt.xlabel('Market Condition Index')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
