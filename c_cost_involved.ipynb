```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simulate marketing costs
def calculate_marketing_costs(n_simulations, mean, std):
    return np.random.normal(mean, std, n_simulations)
```

```python
# Example usage
n_simulations = 1000
marketing_costs = calculate_marketing_costs(n_simulations, mean=5000, std=1000)
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.hist(marketing_costs, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Marketing Costs')
plt.xlabel('Cost ($)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Simulate legal costs
def calculate_legal_costs(n_simulations, min_val, mode_val, max_val):
    return np.random.triangular(min_val, mode_val, max_val, n_simulations)
```

```python
# Example usage
legal_costs = calculate_legal_costs(n_simulations, min_val=1000, mode_val=2000, max_val=4000)
```

```python
plt.figure(figsize=(10, 6))
plt.hist(legal_costs, bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribution of Legal Costs')
plt.xlabel('Cost ($)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
