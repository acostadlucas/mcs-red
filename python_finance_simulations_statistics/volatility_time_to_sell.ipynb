```python
import numpy as np
```

```python
# Parámetros del estudio
prob_facil = 0.15
prob_igual = 0.58  
prob_dificil = 0.27
```

```python
# Tiempos asumidos (en meses)
tiempo_facil = 6
tiempo_igual = 8
tiempo_dificil = 12
```

```python
# Generar 10,000 observaciones sintéticas
np.random.seed(42)
n_sims = 10000
```

```python
muestra = np.random.choice(
    [tiempo_facil, tiempo_igual, tiempo_dificil],
    size=n_sims,
    p=[prob_facil, prob_igual, prob_dificil]
)
```

```python
# Calcular estadísticos
media = np.mean(muestra)
log_muestra = np.log(muestra)
volatilidad = np.std(log_muestra)
```

```python
print(f"Tiempo promedio: {media:.2f} meses")
print(f"Volatilidad (σ): {volatilidad:.3f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualización: histograma de la muestra
plt.figure(figsize=(10, 6))
sns.histplot(muestra, bins=[5, 6.5, 8.5, 13], kde=False, color='skyblue', edgecolor='black')
plt.xticks([6, 8, 12])
plt.xlabel('Tiempo estimado (meses)')
plt.ylabel('Frecuencia')
plt.title('Distribución de tiempos en la muestra sintética')
plt.grid(True)
plt.tight_layout()
plt.show()
```
